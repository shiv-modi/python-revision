```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,200)
```

```python
y = np.sin(x)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y)
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
colors = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y,c=colors,s=sizes,alpha=.6)
plt.xlabel("This is X  Data")
plt.ylabel("This is Y Data")
plt.title("This is my title")
```

```python
x = ['a','b','c','d','e']
```

```python
y = np.random.rand(5)
plt.barh(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title("Bar Plot")
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,'--r')
plt.xlabel("this is my x data")
plt.ylabel("this is my y data")
plt.title("this is my title")
plt.grid()
plt.show()  
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x,y,c='b',s=sizes,alpha=.6,marker='v')
plt.xlabel("This is X  Data")
plt.ylabel("This is Y Data")
plt.title("This is my title")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```
